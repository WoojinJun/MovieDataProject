Checking Runtime vs. Reivew Scores

```python
runtime_df = original_rating_df.copy()
runtime_df = runtime_df.drop(["movie_id", "person_id", "start_year", "birth_year"], axis = 1)
runtime_df = runtime_df.drop_duplicates()
runtime_df = runtime_df.reset_index()
runtime_df['Movie Title'] = runtime_df['primary_title']
runtime_df['Director'] = runtime_df['primary_name']
runtime_df['Rating'] = runtime_df['averagerating']
runtime_df['NumVotes'] = runtime_df['numvotes']
runtime_df['Genres'] = runtime_df['genres']
cols = ['Movie Title','Director','Rating','NumVotes','Genres','primary_profession','death_year', "runtime_minutes"]
runtime_df = runtime_df[cols]
plt.scatter(x = runtime_df['runtime_minutes'], y = runtime_df['Rating'])

plt.title("Runtime vs. Review Scores")
plt.xlabel("Movie Runtime(in min)")
plt.ylabel("Review Scores")
```

```python
filtered_runtime = runtime_df[runtime_df['runtime_minutes'] < 600]
plt.scatter(x = filtered_runtime['runtime_minutes'], y = filtered_runtime['Rating'])
plt.title("Runtime vs. Review Scores Outliers Filtered")
plt.xlabel("Movie Runtime(in min)")
plt.ylabel("Review Scores")
```
